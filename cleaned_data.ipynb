```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
data = pd.read_csv('water_use.txt', error_bad_lines=False, sep = "\t", header = 0)
```

```python
data
```

```python
data = data.drop(index=0)
```

```python
data = data.transpose()
```

```python
data
```

```python
data = data.replace({'c': np.nan})
```

```python
data = data.replace({'b': 0})
```

```python
data = data.applymap(lambda x: x.strip("a") if isinstance(x, str) else x)
```

```python
data = data.reset_index()
```

```python
data = data.rename(columns = {'index':'year', 1:'population', 2:'total_withdrawals', 3:'public_supply', 4:'self_supplied_domestic', 5:'livestock', 6:'irrigation', 7:'thermoelectric_power', 8:'self_supplied_industrial', 9:'mining', 10:'commercial', 11:'aquacultrue', 12:'total_groundwater_fresh', 13:'total_groundwater_saline', 14:'total_surface_water_fresh', 15:'total_surface_water_saline'})
```

```python
data
```

```python
data = data.drop(index=0)
```

```python
data
```

```python
data["year"] = pd.to_datetime(data["year"], errors='coerce').dt.year
```

```python
data["population"] = data["population"].astype('float')
```

```python
data["total_withdrawals"] = data["total_withdrawals"].astype('int64')
```

```python
data["public_supply"] = data["public_supply"].astype('float')
```

```python
data["self_supplied_domestic"] = data["self_supplied_domestic"].astype('float')
```

```python
data["livestock"] = data["livestock"].astype('float')
```

```python
data["irrigation"] = data["irrigation"].astype('int64')
```

```python
data["thermoelectric_power"] = data["thermoelectric_power"].astype('int64')
```

```python
data["self_supplied_industrial"] = data["self_supplied_industrial"].astype('float')
```

```python
data["total_groundwater_fresh"] = data["total_groundwater_fresh"].astype('float')
```

```python
data["total_groundwater_saline"] = data["total_groundwater_saline"].astype('float')
```

```python
data["total_surface_water_fresh"] = data["total_surface_water_fresh"].astype('int64')
```

```python
data["total_surface_water_saline"] = data["total_surface_water_saline"].astype('float')
```
